```python
# ── IMPORTS (à lancer en premier) ──
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('Librairies chargées ✓')
```

## Section 1 — Import / Export

```python
data = {
    'produit':       ['Stylo','Cahier','Règle','Gomme','Crayon',
                      'Stylo','Cahier','Règle','Gomme','Crayon'],
    'categorie':     ['Ecriture','Papeterie','Mesure','Ecriture','Ecriture',
                      'Ecriture','Papeterie','Mesure','Ecriture','Ecriture'],
    'quantite':      [10, 25, 5, 30, 20, 15, 10, 8, 45, 12],
    'prix_unitaire': [1.5, 3.0, 2.5, 0.5, 0.8, 1.5, 3.0, 2.5, 0.5, 0.8],
    'ville':         ['Paris','Lyon','Paris','Marseille','Lyon',
                      'Paris','Paris','Lyon','Marseille','Paris'],
    'mois':          ['Jan','Jan','Jan','Jan','Jan',
                      'Fev','Fev','Fev','Fev','Fev']
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('ventes.csv', index=False)       # Export CSV
df.to_excel('ventes.xlsx', index=False)    # Export Excel
df.to_json('ventes.json')                  # Export JSON
print('Fichiers exportés : ventes.csv, ventes.xlsx, ventes.json')
```

## Section 4 — Nettoyage

```python
df_dirty = df.copy()
df_dirty.loc[2, 'quantite']      = None
df_dirty.loc[5, 'prix_unitaire'] = None
df_dirty.loc[8, 'ville']         = None
df_dirty = pd.concat([df_dirty, df_dirty.iloc[[0, 1]]], ignore_index=True)
print('Valeurs nulles :') ; display(df_dirty.isnull().sum())
```

```python
df_dirty['quantite']      = df_dirty['quantite'].fillna(df_dirty['quantite'].median())
df_dirty['prix_unitaire'] = df_dirty['prix_unitaire'].fillna(0)
df_dirty = df_dirty.dropna(subset=['ville'])
df_dirty = df_dirty.drop_duplicates()
df_dirty = df_dirty.replace('Paris', 'IDF')
df_dirty = df_dirty.rename(columns={'prix_unitaire': 'prix'})
df_dirty['quantite'] = df_dirty['quantite'].astype(int)
df_dirty = df_dirty.reset_index(drop=True)
df_dirty
```

## Section 5 — Tri & Filtrage

```python
df['total_vente'] = df['quantite'] * df['prix_unitaire']
df.head()
```

```python
print('sort ASC :') ; display(df.sort_values('total_vente').head(3))
print('sort DESC :') ; display(df.sort_values('total_vente', ascending=False).head(3))
```

```python
print('sort multi-colonnes :') ; display(df.sort_values(['ville','total_vente'], ascending=[True,False]).head(4))
print('query prix < 1.0 :') ; display(df.query('prix_unitaire < 1.0'))
```

```python
print('nlargest(3) :') ; display(df.nlargest(3, 'total_vente'))
print('nsmallest(3) :') ; display(df.nsmallest(3, 'total_vente'))
print('filter like=\'e\' :') ; display(df.filter(like='e').head(3))
```

## Section 6 — Groupement & Agrégation

```python
print('groupby sum par catégorie :') ; display(df.groupby('categorie')['quantite'].sum())
print('groupby mean par produit :') ; display(df.groupby('produit')['prix_unitaire'].mean())
```

```python
# Pivot table
df.pivot_table(values='total_vente', index='ville', columns='mois', aggfunc='sum')
```

```python
print('agg multi :') ; display(df.agg({'quantite': 'sum', 'prix_unitaire': 'mean'}))
print('apply mean :') ; display(df[['quantite','prix_unitaire','total_vente']].apply(np.mean))
```

```python
# transform : pourcentage du total
df['pct_total'] = df['total_vente'].transform(lambda x: round(x / x.sum() * 100, 2))
df[['produit', 'total_vente', 'pct_total']]
```

## Section 7 — Opérations Statistiques

```python
cols = ['quantite', 'prix_unitaire', 'total_vente']
stats = pd.DataFrame({
    'mean'   : df[cols].mean(),
    'median' : df[cols].median(),
    'std'    : df[cols].std(),
    'var'    : df[cols].var(),
    'min'    : df[cols].min(),
    'max'    : df[cols].max(),
    'sum'    : df[cols].sum(),
})
stats
```

```python
# Matrice de corrélation
df[['quantite', 'prix_unitaire', 'total_vente']].corr()
```

## Section 8 — Visualisation

```python
# Bar chart — ventes par produit
df.groupby('produit')['total_vente'].sum().plot(kind='bar', color='steelblue', title='Ventes par produit')
plt.tight_layout() ; plt.show()
```

```python
# Barres horizontales
df.groupby('produit')['total_vente'].sum().plot(kind='barh', color='coral', title='Barres horizontales')
plt.tight_layout() ; plt.show()
```

```python
# Histogramme
df['total_vente'].plot(kind='hist', bins=6, color='green', title='Distribution total_vente')
plt.tight_layout() ; plt.show()
```

```python
# Box plot
df[['quantite', 'total_vente']].plot(kind='box', title='Box plot')
plt.tight_layout() ; plt.show()
```

```python
# KDE
df['quantite'].plot(kind='kde', title='KDE quantite')
plt.tight_layout() ; plt.show()
```

```python
# Pie chart
df.groupby('ville')['total_vente'].sum().plot(kind='pie', autopct='%1.0f%%', title='Répartition par ville')
plt.tight_layout() ; plt.show()
```

```python
# Scatter
df.plot.scatter(x='quantite', y='total_vente', color='purple', title='Scatter quantite vs total')
plt.tight_layout() ; plt.show()
```

```python
# Line — évolution mensuelle
df.groupby('mois')['total_vente'].sum().plot(kind='line', marker='o', title='Évolution mensuelle')
plt.tight_layout() ; plt.show()
```

```python
# Area plot
df.groupby('produit')['total_vente'].sum().plot(kind='area', alpha=0.5, title='Area plot ventes')
plt.tight_layout() ; plt.show()
```

```python
# Tous les graphiques en une grille
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('TP Pandas — Toutes les visualisations', fontsize=16, fontweight='bold')

df.groupby('produit')['total_vente'].sum().plot(kind='bar',  ax=axes[0,0], title='Bar', color='steelblue')
df.groupby('produit')['total_vente'].sum().plot(kind='barh', ax=axes[0,1], title='Barh', color='coral')
df['total_vente'].plot(kind='hist', ax=axes[0,2], title='Hist', bins=6, color='green')
df[['quantite','total_vente']].plot(kind='box', ax=axes[1,0], title='Box')
df['quantite'].plot(kind='kde', ax=axes[1,1], title='KDE')
df.groupby('ville')['total_vente'].sum().plot(kind='pie', ax=axes[1,2], title='Pie', autopct='%1.0f%%')
df.plot.scatter(x='quantite', y='total_vente', ax=axes[2,0], title='Scatter', color='purple')
df.groupby('mois')['total_vente'].sum().plot(kind='line', ax=axes[2,1], title='Line', marker='o')
df.groupby('produit')['total_vente'].sum().plot(kind='area', ax=axes[2,2], title='Area', alpha=0.5)

plt.tight_layout()
plt.savefig('visualisations.png', dpi=120, bbox_inches='tight')
plt.show()
print('Grille sauvegardée : visualisations.png')
```
